```python
import matplotlib.pyplot as plt
import seaborn as sns
import plotly as py
import plotly.graph_objs as go
from sklearn.cluster import KMeans
import warnings
import os
```

```python
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Projetos DS/Agrupamento de Clientes/Mall_Customers.csv')
```

```python
df.head(10)
```

```python
characters = list(df.columns)
characters.remove('CustomerID')
characters.remove('Gender')
```

```python
#Gráfico de distribuição das variáveis

plt.figure(1, figsize = (15, 6))
n = 0

for i in characters:
  n += 1
  plt.subplot(1, 3, n)
  plt.subplots_adjust(hspace = 0.5, wspace = 0.5)
  sns.distplot(df[i], bins = 25)
  plt.title('{}'.format(i))

plt.show()
```

```python
#Relação entre atributos
plt.figure(1, figsize = (15,6))

for gender in ['Male', 'Female']:
  plt.scatter( x = 'Age', y = 'Annual Income (k$)' , data = df[df['Gender'] == gender],
              s = 100, alpha = 0.5, label = gender)


plt.xlabel('Age')
plt.ylabel('Annual Income (k$)')
plt.legend()
plt.show()
```

```python
#Verificando RendaXGasto
plt.figure(1, figsize = (15, 6))

for gender in ['Male', 'Female']:
  plt.scatter( x = 'Annual Income (k$)', y ='Spending Score (1-100)' , data = df[df['Gender'] == gender],
              s = 100, alpha = 0.5, label = gender)


plt.ylabel('Spending Score')
plt.xlabel('Annual Income (k$)')
plt.title('Renda Anual vs Pontuação de Gastos')
plt.legend()
plt.show()
```
